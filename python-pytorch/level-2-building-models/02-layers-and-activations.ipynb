# Layers and Activation Functions

Layers = linear transformations (matrix math)
Activations = non-linearities applied after layers

Without activations, stacking 10 linear layers = still just one linear layer.
Activations let the network learn non-linear patterns (which is everything real).

```python
import torch
import torch.nn as nn

# --- Core Layers ---

# nn.Linear — fully connected layer: y = xW^T + b
linear = nn.Linear(in_features=8, out_features=4)
x = torch.randn(2, 8)   # batch=2, features=8
print('Linear output shape:', linear(x).shape)   # [2, 4]

# nn.Embedding — lookup table: integer index → dense vector
# Used for word tokens, user IDs, any categorical data
embed = nn.Embedding(num_embeddings=1000, embedding_dim=64)  # vocab=1000, embed_dim=64
tokens = torch.tensor([1, 5, 200, 999])   # token indices
print('Embedding output shape:', embed(tokens).shape)   # [4, 64]

# nn.LayerNorm — normalizes across features (used in Transformers)
layer_norm = nn.LayerNorm(64)
print('LayerNorm output shape:', layer_norm(embed(tokens)).shape)   # [4, 64]

# nn.Dropout — randomly zeros neurons during training (regularization)
dropout = nn.Dropout(p=0.1)   # 10% dropout
print('Dropout output shape:', dropout(embed(tokens)).shape)   # [4, 64]
```

## Activation Functions — the non-linearities

Applied element-wise after linear layers. Each has a different shape and use case.

```python
x = torch.tensor([-2.0, -1.0, 0.0, 1.0, 2.0])

# ReLU — max(0, x) — most common in CNNs and MLPs
# Dead neuron problem: if x<0, gradient=0, neuron never updates
print('ReLU:    ', torch.relu(x))

# GELU — smooth version of ReLU — standard in Transformers (BERT, GPT)
# slightly better than ReLU for deep networks
print('GELU:    ', torch.nn.functional.gelu(x))

# SiLU (Swish) — x * sigmoid(x) — used in LLaMA, Mistral
print('SiLU:    ', torch.nn.functional.silu(x))

# Sigmoid — squashes to (0,1) — used in binary classification output
# Not used in hidden layers (vanishing gradient problem)
print('Sigmoid: ', torch.sigmoid(x))

# Tanh — squashes to (-1,1) — used in RNNs/LSTMs
print('Tanh:    ', torch.tanh(x))
```

## Activation cheat sheet

| Activation | Formula | Use case |
|---|---|---|
| ReLU | max(0, x) | CNN hidden layers, MLPs |
| GELU | x * Φ(x) | BERT, GPT-2, Transformers |
| SiLU/Swish | x * σ(x) | LLaMA, Mistral, modern LLMs |
| Sigmoid | 1/(1+e^-x) | Binary classification output only |
| Tanh | (e^x-e^-x)/(e^x+e^-x) | RNN/LSTM gates |
| Softmax | e^xi / Σe^xj | Multi-class output, attention weights |

```python
# Why activations matter — with vs without

# Without activation: 3 linear layers = still just 1 linear layer
# (linear composed with linear is still linear)
no_activation = nn.Sequential(
    nn.Linear(2, 4),
    nn.Linear(4, 4),
    nn.Linear(4, 1)
)

# With activation: can learn non-linear boundaries
with_activation = nn.Sequential(
    nn.Linear(2, 4),
    nn.ReLU(),
    nn.Linear(4, 4),
    nn.ReLU(),
    nn.Linear(4, 1)
)

x = torch.randn(5, 2)
print('Without activation output shape:', no_activation(x).shape)
print('With activation output shape:   ', with_activation(x).shape)

# Both produce same shape, but with_activation can approximate any function
# no_activation can only fit linear relationships
```
